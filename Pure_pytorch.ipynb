```python
from fastai.vision import *
from fastai import *
```

```python
data = ImageDataBunch.from_folder('./data/chest_xray/', valid='val', size=224, bs=16,num_workers=0)
```

```python
data.show_batch(rows=3, figsize=(10,10))
```

```python
learn = cnn_learner(data, models.resnet50, metrics=[accuracy], callback_fns= ShowGraph)
```

```python
learn.lr_find()
learn.recoder.plot()
```
